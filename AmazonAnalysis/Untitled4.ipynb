```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of DataFrames and corresponding categories
category_data = {"Shoes": Shoes,
    "Furniture": Furniture,
    "Electronics": Electronics
}

# Process each category's DataFrame
for category, df in category_data.items():
    df["review_date"] = pd.to_datetime(df["review_date"], errors="coerce")  # Convert to datetime
    df.dropna(subset=["review_date", "star_rating"], inplace=True)  # Remove rows with missing values
    df["year"] = df["review_date"].dt.year  # Extract year
    category_data[category] = df.groupby("year")["star_rating"].mean().reset_index()  # Calculate average rating per year

# Plot the trends
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

for category, data in category_data.items():
    sns.lineplot(x=data["year"], y=data["star_rating"], label=category)
sns.set_context("talk")
plt.xlabel("Year")
plt.ylabel("Average Star Rating")
plt.title("Average Star Rating Over Time by Product Category")
plt.legend(title="Product Category", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()

```
