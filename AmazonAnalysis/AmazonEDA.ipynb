```python
#!pip install vaderSentiment # Install the vaderSentiment package
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
analyzer = SentimentIntensityAnalyzer()
df= Electronics

      # Sentiment Analysis
#df['sentiment'] = df['review_body'].apply(lambda x: analyzer.polarity_scores(str(x))['compound'])

sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(x='review_date', y='sentiment', data=df, alpha=0.5)
plt.xlabel("Star Rating")
plt.ylabel("Sentiment Polarity")
plt.title("Sentiment vs. Star Rating")
plt.show()
```

```python
df['review_date'] = pd.to_datetime(df['review_date'])
sns.set_context("talk")

sentiment_trend = df.groupby('review_date')['sentiment'].mean().reset_index()

df['review_year'] = df['review_date'].dt.year

sentiment_by_year = df.groupby('review_year')['sentiment'].mean().reset_index()

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x='review_year', y='sentiment', data=sentiment_by_year, marker="o")

plt.xlabel("Year")
plt.ylabel("Average Sentiment Polarity")
plt.title("Average Sentiment Over Time by Year for Electronics")

plt.show()
```

```python
# Group by year and count the number of reviews
sns.set_context("talk")
reviews_by_year = df.groupby('review_year').size().reset_index(name='num_reviews')

# Plot the number of reviews over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='review_year', y='num_reviews', data=reviews_by_year, marker="o")

plt.xlabel("Year")
plt.ylabel("Number of Reviews")
plt.title("Number of Reviews Over Time for Electronics")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
avg_rating_by_year = df.groupby('review_year')['star_rating'].mean().reset_index()

# Set font size for the plot
sns.set_context("talk")  # This increases the font size for labels, ticks, and titles

# For more control, you can adjust the font size individually for each element
plt.figure(figsize=(10, 6))
sns.lineplot(x='review_year', y='star_rating', data=avg_rating_by_year, marker="o")
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Star Rating", fontsize=14)
plt.title("Average Star Rating Over Time for Electronics", fontsize=16)

plt.show()
```

```python

filtered_data = df[df['review_year'].between(2012, 2014)]

top_products_per_year = filtered_data.groupby(['review_year', 'product_title'])['star_rating'].mean().reset_index()
top_10_products_per_year = top_products_per_year.sort_values(['review_year', 'star_rating'], ascending=[True, False])

top_10_per_year = top_10_products_per_year.groupby('review_year').head(10).reset_index(drop=True)
print(top_10_per_year[['review_year','product_title', 'star_rating']])
```
